```python
# finding the upper and lower limit of horespower and acceleration columns

def find_outlier_limits(col_name):
    Q1,Q3=wine_df[col_name].quantile([.25,.75])
    IQR=Q3-Q1
    low=Q1-(1.5* IQR)
    high=Q3+(1.5* IQR)
    return (high,low)

high_fix,low_fix=find_outlier_limits('fixed acidity')
print('fixed acidity: ','upper limit: ',high_fix,' lower limit: ',low_fix)
high_va,low_va=find_outlier_limits('volatile acidity')
print('volatile acidity: ','upper limit: ',high_va,' lower limit:',low_va)
high_ca,low_ca=find_outlier_limits('citric acid')
print('citric acid: ','upper limit: ',high_ca,' lower limit: ',low_ca)
high_rs,low_rs=find_outlier_limits('residual sugar')
print('residual sugar: ','upper limit: ',high_rs,' lower limit: ',low_rs)
high_chl,low_chl=find_outlier_limits('chlorides')
print('chlorides: ','upper limit: ',high_chl,' lower limit: ',low_chl)
high_fsd,low_fsd=find_outlier_limits('free sulfur dioxide')
print('free sulfur dioxide: ','upper limit: ',high_fsd,' lower limit: ',low_fsd)
high_tsd,low_tsd=find_outlier_limits('total sulfur dioxide')
print('total sulfur dioxide: ','upper limit: ',high_tsd,' lower limit: ',low_tsd)
high_den,low_den=find_outlier_limits('density')
print('density: ','upper limit: ',high_den,' lower limit: ',low_den)
high_pH,low_pH=find_outlier_limits('pH')
print('pH: ','upper limit: ',high_pH,' lower limit: ',low_pH)
high_sulp,low_sulp=find_outlier_limits('sulphates')
print('sulphates: ','upper limit: ',high_sulp,' lower limit: ',low_sulp)
high_alc,low_alc=find_outlier_limits('alcohol')
print('alcohol: ','upper limit: ',high_alc,' lower limit: ',low_alc)
high_qu,low_qu=find_outlier_limits('quality')
print('quality: ','upper limit: ',high_qu,' lower limit: ',low_qu)


# Replacing outlier values in horespower and acceleration columns with respective 
# upper and lower limits

wine_df.loc[wine_df['fixed acidity']>high_fix,'fixed acidity']=high_fix
wine_df.loc[wine_df['volatile acidity']>high_va,'volatile acidity']=high_va
wine_df.loc[wine_df['citric acid']>high_ca,'citric acid']=high_ca
wine_df.loc[wine_df['residual sugar']>high_rs,'residual sugar']=high_rs
wine_df.loc[wine_df['residual sugar']<low_rs,'residual sugar']=low_rs
wine_df.loc[wine_df['chlorides']>high_chl,'chlorides']=high_chl
wine_df.loc[wine_df['chlorides']<low_chl,'chlorides']=low_chl
wine_df.loc[wine_df['free sulfur dioxide']>high_fsd,'free sulfur dioxide']=high_fsd
wine_df.loc[wine_df['total sulfur dioxide']>high_tsd,'total sulfur dioxide']=high_tsd
wine_df.loc[wine_df['density']>high_den,'density']=high_den
wine_df.loc[wine_df['density']<low_den,'density']=low_den
wine_df.loc[wine_df['pH']>high_pH,'pH']=high_pH
wine_df.loc[wine_df['sulphates']>high_sulp,'sulphates']=high_sulp
wine_df.loc[wine_df['alcohol']>high_alc,'alcohol']=high_alc
wine_df.loc[wine_df['quality']>high_qu,'quality']=high_qu
wine_df.loc[wine_df['quality']<low_qu,'quality']=low_qu
```

```python
# dividing the dataset into dependent and independent variables

x = wine_df.iloc[:,:11]
y = wine_df.iloc[:,11]

# determining the shape of x and y.
print(x.shape)
print(y.shape)
```

```python
# dividing the dataset in training and testing set

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 44)

# determining the shapes of training and testing sets
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# standard scaling 
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

### Modelling

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

from sklearn.model_selection import GridSearchCV, cross_val_score



# creating the model
model = LogisticRegression()

# feeding the training set into the model
model.fit(x_train, y_train)

# predicting the results for the test set
y_pred = model.predict(x_test)

# calculating the training and testing accuracies
print("Training accuracy :", model.score(x_train, y_train))
print("Testing accuracy :", model.score(x_test, y_test))

# classification report
print(classification_report(y_test, y_pred))

# confusion matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# code
x_pred = list(model.predict(x_test))
df = {'predicted':x_pred,'original':y_test}
pd.DataFrame(df).head(10)
```
